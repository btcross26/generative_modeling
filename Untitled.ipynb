```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs

import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
```

```python
# sample data
X, c = make_blobs(
    n_samples=10000,
    n_features=64,
    centers=10,
    random_state=5,
)
```

```python
mu = X.mean(axis=0)
std = X.std(axis=0)
X = (X - mu) / std
```

```python
class PytorchModel(nn.Module):
    def __init__(self, input_size):
        super().__init__()
        self.d1 = nn.Linear(input_size, 32)
        self.r1 = nn.LeakyReLU(0.3)
        self.d2 = nn.Linear(32, 16)
        self.r2 = nn.LeakyReLU(0.3)
        self.d3 = nn.Linear(16, 2)
        
    def inverse_relu(self, X):
        return torch.where(X > 0.0, X, 0.3 * X)
        
    def encode(self, X):
        out = self.d1(X)
        out = self.r1(out)
        out = self.d2(out)
        out = self.r2(out)
        out = self.d3(out)
        return out

    def decode(self, X):
        out = torch.matmul(X - self.d3.bias.unsqueeze(0), self.d3.weight)
        out = self.inverse_relu(out)
        out = torch.matmul(out - self.d2.bias.unsqueeze(0), self.d2.weight)
        out = self.inverse_relu(out)
        out = torch.matmul(out - self.d1.bias.unsqueeze(0), self.d1.weight)
        return out
    
    def forward(self, X):
        out = self.encode(X)
        out = self.decode(X)
        return out
    
    def fit(self, X, learning_rate=0.0005, vae_alpha=0.001, batch_size=32, epochs=10, shuffle=True):
        optimizer = torch.optim.Adam(lr=learning_rate, params=self.parameters())
        dataset = TensorDataset(X)
        train_loader = torch.utils.data.DataLoader(
            dataset,
            batch_size=batch_size,
            shuffle=shuffle,
        )

        for epoch in range(1, epochs + 1):
            n, total_loss = 0, 0.0
            for obs, in train_loader:
                out = self.encode(obs)
                mle_loss = self._normal_mle_loss(out)
                out = self.decode(out)
                ls_loss = self._ls_loss(out, obs)
                loss = vae_alpha * mle_loss + (1.0 - vae_alpha) * ls_loss
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
                
                n += out.shape[0]
                total_loss += float(loss.detach().item()) * obs.shape[0]
                
            avg_loss = 1.0 * total_loss / n
            print(f"Epoch {epoch} Loss: {vae_alpha * mle_loss:.6f}, {(1.0 - vae_alpha) * ls_loss:.6f}, {avg_loss:.6f}")
            
        return self

    def _normal_mle_loss(self, out):
        kl_loss = torch.sum(out ** 2, axis=1)
        return torch.mean(kl_loss)
    
    def _ls_loss(self, out, obs):
        ls_loss = torch.mean((out - obs) ** 2, axis=[1])
        return torch.mean(ls_loss)
    
    def _initialize_weights(self):
        with torch.no_grad():
            self.apply(self.init_weights)
                
    @staticmethod
    def init_weights(layer):
        if type(layer) in [nn.BatchNorm1d, nn.BatchNorm2d]:
            layer.weight.data.fill_(1.0)
            layer.bias.data.fill_(0.0)
        else:
            if hasattr(layer, "weight"):
                torch.nn.init.xavier_uniform_(layer.weight)

            if hasattr(layer, "bias"):
                layer.bias.data.fill_(0.0)
```

```python
model = PytorchModel(X.shape[1])
```

```python
# could try clustering metrics as a surrogate loss function...
# also could mess with learning rate scheduler...
model.fit(
    torch.from_numpy(X).to(dtype=torch.float32),
    epochs=200,
    vae_alpha=0.01,
)
```

```python
model.eval()
with torch.no_grad():
    pts = model.encode(torch.from_numpy(X).type(torch.float32))
```
